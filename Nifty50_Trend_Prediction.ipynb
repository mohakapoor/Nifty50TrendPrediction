# Importing Libraries & Data

```python
import yfinance as yf
import pandas as pd
import pandas as pd
```

```python
nifty50 = yf.Ticker("^NSEI")
```

```python
nifty50 = nifty50.history(period="max")
```

```python
nifty50
```

```python
nifty50.plot.line(y = "Close", use_index=True)
```

# Feature Engineering

```python
del nifty50["Dividends"]
del nifty50["Stock Splits"]
```

```python
nifty50["Tomorrow"] = nifty50["Close"].shift(-1)
```

```python
nifty50
```

```python
nifty50["Target"] = (nifty50["Tomorrow"]>nifty50["Close"]).astype(int)
nifty50
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = nifty50.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    nifty50[ratio_column] = nifty50["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    nifty50[trend_column] = nifty50.shift(1).rolling(horizon).sum()["Target"]

    new_predictors+= [ratio_column, trend_column]
```

```python
nifty50
```

```python
nifty50 = nifty50.dropna(subset = nifty50.columns[nifty50.columns != "Tomorrow"])
```

```python
nifty50
```

# Building Our Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 200, min_samples_split =50,random_state = 1)
```

```python
def predict(train,test,predictors,model):
  model.fit(train[predictors],train["Target"])
  preds = model.predict_proba(test[predictors])[:,1]
  preds[preds>= .6] = 1
  preds[preds < .6] = 0
  preds = pd.Series(preds, index = test.index, name= "Predictions")
  return pd.concat([test["Target"],preds],axis = 1)
```

```python
def backtest(data, model, predictors, start = 1250, step =250):
  all_predictions = []
  for i in range(start, data.shape[0],step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = predict(train,test,predictors,model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```

# Predicitons & Visualizations

```python
predictions = backtest(nifty50,model,new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
from sklearn.metrics import precision_score,confusion_matrix
Prec_score = precision_score(predictions["Target"],predictions["Predictions"])
Prec_score = float(f"{Prec_score:.2f}")
```

```python
market_up = (predictions["Target"].value_counts())[1]/predictions.shape[0]
market_up = float(f"{market_up:.2f}")
```

```python
import matplotlib.pyplot as plt

categories = ['Model Accuracy', '% of time Stock has gone up']
percentages = [Prec_score, market_up]  # Example: 70% and 30%

plt.bar(categories, percentages, color=['blue', 'orange'])

for i, percentage in enumerate(percentages):
    plt.text(i, 0.67 , f'{percentage}%', ha='center', fontsize=12)
plt.show()
```
